```python
df.dtypes
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Cut(Quality)'] = label_encoder.fit_transform(df['Cut(Quality)'])
df['Cut(Quality)'].unique()
```

```python
df['Color'] = label_encoder.fit_transform(df['Color'])
df['Color'].unique()
```

```python
df['Clarity'] = label_encoder.fit_transform(df['Clarity'])
df['Clarity'].unique()
```

```python
#model
x = df.drop("Price(in US dollars)", axis=1)
y = df["Price(in US dollars)"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
```

# **Decision Tree Regressor**

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(random_state=0)
dtree.fit(x_train, y_train)
```

Calculating the loss after training

```python
from sklearn import metrics
import math
y_pred = dtree.predict(x_test)
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
rmse = math.sqrt(mse)

print('MAE is {}'.format(mae))
print('MSE is {}'.format(mse))
print('r2 is {}'.format(r2))
print('RMSE is {}'.format(rmse))
```

```python
dec_score = dtree.score(x_train,y_train)
print(dec_score)
```

```python
dtr_score = dtree.score(x_test,y_test)
dtr_score
```

```python
from sklearn.tree import export_graphviz

# export the decision tree model to a tree_structure.dot file
# paste the contents of the file to webgraphviz.com
export_graphviz(dtree, out_file ='tree_structure.dot',
               feature_names =[car, cut, co, cl, de, ta, x, y, z])
```

# **RANDOM FOREST REGRESSOR**

```python
#RANDOM FOREST REGRESSOR
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state=0)
rf.fit(x_train, y_train)
```

```python
from sklearn import metrics
import math
y_pred = rf.predict(x_test)
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)
rmse = math.sqrt(mse)

print('MAE is {}'.format(mae))
print('MSE is {}'.format(mse))
print('r2 is {}'.format(r2))
print('RMSE is {}'.format(rmse))
```

```python
rs = rf.score(x_train,y_train)
print(rs)
```

```python
rs_score = rf.score(x_test,y_test)
print(rs_score)
```
